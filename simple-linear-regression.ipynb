# We read the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data=pd.read_csv('/kaggle/input/linear-regression-dataset/Linear Regression - Sheet1.csv')
```

# Data Cleaning

```python
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
imp=imp.fit(data)
data=imp.transform(data)
```

# Split the data into input and output

```python
X=data[ : , 0]
# X is the input
y=data[ : , 1]
#y is the output
```

# Data Scaling

```python
st=StandardScaler(copy=True,with_mean=True,with_std=True)
newdata=st.fit_transform(data)
newdata
```

# Data Spliting

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=33)
```

# Predicition

```python
lr=LinearRegression()
lr.fit(X_train.reshape(-1,1),y_train.reshape(-1,1))
lr.score(X_train.reshape(-1,1),y_train.reshape(-1,1))
lr.score(X_test.reshape(-1,1),y_test.reshape(-1,1))
y_pred=lr.predict(X_test.reshape(-1,1))
```

# Calculating the error value

```python
mean_absolute_error(y_test.reshape(-1,1), y_pred.reshape(-1,1))
```
